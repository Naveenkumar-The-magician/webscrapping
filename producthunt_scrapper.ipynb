```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

# browser.get("https://www.producthunt.com/@tianzhou_chen/followers")

driver = webdriver.Chrome()
browser = driver


user = input("enter the user name:")


class follow:
    def followers(self, user):
        browser.get("https://www.producthunt.com/@" + user + "/followers")
        while True:
            old = len(
                browser.find_elements(
                    By.CSS_SELECTOR,
                    value=".color-darker-grey.fontSize-16",
                )
            )
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            new = len(
                browser.find_elements(
                    By.CSS_SELECTOR,
                    value=".color-darker-grey.fontSize-16",
                )
            )
            if old == new:
                break

        followers = [
            follower.text
            for follower in browser.find_elements(
                By.CSS_SELECTOR,
                value=".color-darker-grey.fontSize-16",
            )
        ]
        return followers

    def following(self, user):
        browser.get("https://www.producthunt.com/@" + user + "/following")
        while True:
            old = len(
                browser.find_elements(
                    By.CSS_SELECTOR,
                    value=".color-darker-grey.fontSize-16",
                )
            )
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            new = len(
                browser.find_elements(
                    By.CSS_SELECTOR,
                    value=".color-darker-grey.fontSize-16",
                )
            )
            if old == new:
                break

        following = [
            followee.text
            for followee in browser.find_elements(
                By.CSS_SELECTOR,
                value=".color-darker-grey.fontSize-16",
            )
        ]
        return following


bot = follow()


print("followers :\n\n", bot.followers(user))
print("followings :\n\n", bot.following(user))
```
